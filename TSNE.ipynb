```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from numpy import reshape
import seaborn as sns
import pandas as pd
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
tsne= TSNE(n_components=2, verbose=1, random_state=123)
z=tsne.fit_transform(x)
```

```python
df=pd.DataFrame()
df["y"]=y
df["comp_1"]=z[:,0]
df["comp_2"]=z[:,1]
```

```python
sns.scatterplot(x="comp_1", y="comp_2", hue=df.y.tolist(), palette=sns.color_palette("hls",3), data=df).set(title="iris data T-SNE projection")
```
